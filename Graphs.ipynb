```python
import seaborn as sns
sns.__version__
```

```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

```python
automobile_data=pd.read_csv('/Users/prathibha/Downloads/Automobile_data.csv',na_values='?')
automobile_data.head(5)
```

```python
automobile_data.shape
```

```python
automobile_data.isnull().sum()
```

```python
plt.figure(figsize=(12,8))
sns.distplot(automobile_data['price'],color='red',bins=20,hist=True,rug=True)
plt.title('Automobile data')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.rugplot(automobile_data['price'],color='red',height=0.5)
plt.title('Automobile data')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.kdeplot(automobile_data['price'],color='blue',shade=True)
plt.title('Automobile data')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='horsepower',y='price',
                data=automobile_data,s=120,hue='num-of-cylinders')
plt.title('Automobile data')
plt.show()
```

```python
sns.regplot(x='horsepower',y='price',
             data=automobile_data)
```

```python
sns.jointplot(x='horsepower',y='price',
             data=automobile_data)
```
